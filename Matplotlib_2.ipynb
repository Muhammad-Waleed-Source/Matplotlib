```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
iris = pd.read_csv('/content/iris.csv')
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# Now coloring different species

iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})

iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# We can control the colors (Themes) -> viridis, plasma, infreno, magma, cividis, jet etc.

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# We can also check which color represent which value

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()

# Blue dots are in 0 range so they represent "Iris-setosa"
# Green dots are in 1 range so they represent "Iris-versicolor"
# Red dots are in 2 range so they represent "Iris-virginica"
```

```python
# alpha

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet', alpha=0.3)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Plot size

```python
plt.figure(figsize=(15, 7)) # (width, height)

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
plt.figure(figsize=(5, 2))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Annotations

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x, y)

plt.text(1,5, 'Point 1')
plt.text(2,6, 'Point 2')
plt.text(3,7, 'Point 3')
plt.text(4,8, 'Point 4', fontdict={'size':12, 'color':'red'})
```

```python
batters = pd.read_csv('batter.csv')
batters.head(5)
```

```python
sample_df = batters.head(100).sample(25, random_state=5)

sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

### Each dot represent one batsman, but we are unable to see the name, so here we use annotations(placing texts on a graph)

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
# here we can increase the size now
plt.figure(figsize=(16, 8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i], fontdict={'size':12, 'color':'red'})
```
