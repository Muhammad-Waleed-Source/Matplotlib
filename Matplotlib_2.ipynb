```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
iris = pd.read_csv('/content/iris.csv')
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# Now coloring different species

iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})

iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# We can control the colors (Themes) -> viridis, plasma, infreno, magma, cividis, jet etc.

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# We can also check which color represent which value

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()

# Blue dots are in 0 range so they represent "Iris-setosa"
# Green dots are in 1 range so they represent "Iris-versicolor"
# Red dots are in 2 range so they represent "Iris-virginica"
```

```python
# alpha

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet', alpha=0.3)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Plot size

```python
plt.figure(figsize=(15, 7)) # (width, height)

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
plt.figure(figsize=(5, 2))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
batters = pd.read_csv('batter.csv')
batters.head(5)
```

```python
sample_df = batters.head(100).sample(25, random_state=5)

sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

### Each dot represent one batsman, but we are unable to see the name, so here we use annotations(placing texts on a graph)

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
# here we can increase the size now
plt.figure(figsize=(16, 8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i], fontdict={'size':12, 'color':'red'})
```

### Size

```python
# Bubble size based on runs scored by batter

plt.figure(figsize=(16, 8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i], fontdict={'size':12, 'color':'red'})
```

### Horizontal and Vertical lines

```python
# batters with strike rate higher than 130

plt.figure(figsize=(16, 8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs']*0.1)

plt.axhline(130, color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i], fontdict={'size':12, 'color':'red'})

plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# batters with strike rate higher than 130 and also avg greater than 30

plt.figure(figsize=(16, 8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs']*0.1)

plt.axhline(130, color='red')
plt.axvline(30, color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i], fontdict={'size':12, 'color':'red'})

plt.xlabel('Average')
plt.ylabel('Strike Rate')

# So the top right corner represent our desire data.
```

```python
# Strike rate higher than 140 and avg higher than 30 (Exceptional Hitters)

plt.figure(figsize=(16, 8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs']*0.1)

plt.axhline(130, color='red')
plt.axhline(140, color='red')
plt.axvline(30, color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i], fontdict={'size':12, 'color':'red'})

plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

### Subplots

```python
batters.head()
```

```python
# 1. First way to plot a graph

plt.figure(figsize=(15, 6))
plt.scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')

plt.title('Avg Strike')

plt.xlabel('Avg')
plt.ylabel('Strike rate')

plt.show() # to avoid technical info
```

```python
# 2. Second way to plot a graph (Used for multiple plots)

fig, ax = plt.subplots(figsize=(15, 6))

ax.scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax.set_title('Avg Strike')

ax.set_xlabel('Avg')
ax.set_ylabel('Strike rate')

fig.show() # to avoid technical info
```

```python
# Side by side plotting using subplots

plt.subplots(nrows=2, ncols=1)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 6))

ax[0].scatter(sample_df['avg'], sample_df['strike_rate'], color='red')
ax[1].scatter(sample_df['avg'], sample_df['runs'])

ax[0].set_title('Avg vs Strike rate')
ax[1].set_title('Avg vs Runs')


ax[0].set_ylabel('Strike rate')

ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')
```

```python
# 4 subplots

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax[0][0].scatter(sample_df['avg'], sample_df['strike_rate'], color='red')
ax[0][1].scatter(sample_df['avg'], sample_df['runs'])
ax[1][0].hist(sample_df['avg'])
ax[1][1].hist(sample_df['runs'])
```
