### Types of Data

- Numerical Data: Numerical data represents measurable quantities or counts (e.g., temperature, height, number of children) and is analyzed using mathematical and statistical operations.

- Categorical Data: Categorical data describes qualities, characteristics, or group memberships (e.g., car brands, hair color, "yes/no" responses) and is analyzed using groups or proportions.

**Univariate analysis**, **Bivariate analysis** and **Multivariate analysis** are the three techniques are the primary methods for analyzing data based on the number of variables involved. Understanding them is essential for descriptive and inferential statistics.

### 1. Univariate Analysis:
Analyzes a single variable to describe its distribution and identify patterns.

**Example:** Finding the average test score of a class using a histogram.

### 2. Bivariate Analysis:
Compares two variables to determine if a relationship or correlation exists between them.

**Example:** Plotting study hours against exam scores using a scatter plot

### 3. Multivariate Analysis:
Examines three or more variables simultaneously to understand complex interactions and dependencies.

**Example:** Predicting exam scores based on study hours, attendance, and sleep quality.

```python
price = [48000, 54000, 57000, 49000, 47000, 4500000]
year = [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)

# We have an outlier(4500000) in our price data, so our graph will look a odd. The other small values are plotted flat on the graph.
```

```python
# so we can handle this by limiting axis

# limiting y-axis
price = [48000, 54000, 57000, 49000, 47000, 4500000]
year = [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
plt.ylim(0, 75000)
```

```python
# limiting x-axis

price = [48000, 54000, 57000, 49000, 47000, 4500000]
year = [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
plt.ylim(0, 75000)
plt.xlim(2017, 2019)
```

```python
# size
tips = sns.load_dataset('tips')
tips
```

```python
# slower
plt.scatter(tips['total_bill'], tips['tip'])

plt.xlabel('Total Bill')
plt.ylabel('Tip')

# The graph shows a linear relation as bill grows tip also grows.
```

```python
# we can also show some extra information, lets say how many persons(size) are with this person. So the size of the dot will be bigger if he came with many persons and will be smaller if he came with less persons.

plt.scatter(tips['total_bill'], tips['tip'], s=tips['size'])
```

```python
# As the dots are very small because the size col of our dataset contains 2, 4, 1, 5 means small values. So multiplying it with 20 to get a good view.

plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*20)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

```python
# scatter plot using plt.plot (faster)
plt.plot(tips['total_bill'], tips['tip'], 'o')

# using plt.plot we cannot use size attribute as well as we will have problems with color, markers etc. But if we are dealing with very huge data plt.plot is faster.
```
