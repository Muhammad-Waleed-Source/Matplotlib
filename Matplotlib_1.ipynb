### Types of Data

- Numerical Data: Numerical data represents measurable quantities or counts (e.g., temperature, height, number of children) and is analyzed using mathematical and statistical operations.

- Categorical Data: Categorical data describes qualities, characteristics, or group memberships (e.g., car brands, hair color, "yes/no" responses) and is analyzed using groups or proportions.

**Univariate analysis**, **Bivariate analysis** and **Multivariate analysis** are the three techniques are the primary methods for analyzing data based on the number of variables involved. Understanding them is essential for descriptive and inferential statistics.

### 1. Univariate Analysis:
Analyzes a single variable to describe its distribution and identify patterns.

**Example:** Finding the average test score of a class using a histogram.

### 2. Bivariate Analysis:
Compares two variables to determine if a relationship or correlation exists between them.

**Example:** Plotting study hours against exam scores using a scatter plot

### 3. Multivariate Analysis:
Examines three or more variables simultaneously to understand complex interactions and dependencies.

**Example:** Predicting exam scores based on study hours, attendance, and sleep quality.
